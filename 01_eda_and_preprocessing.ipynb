# 01 EDA And Preprocessing

This notebook explores the dataset, compares normalization choices, runs lightweight time-series diagnostics, and saves split/normalization artifacts.

```python
from pathlib import Path

ARTIFACT_ROOT = Path('artifacts')
TRAIN_PATH = Path('Coursework3/viscodata_3mat.mat')
F_FIELD = 'epsi_tol'
SIG_FIELD = 'sigma_tol'
SEED = 20260328
Ntotal = 400
TRAIN_RATIO = 0.70
VAL_RATIO = 0.15
train_size = int(round(Ntotal * TRAIN_RATIO))
val_size = int(round(Ntotal * VAL_RATIO))
test_start = train_size + val_size
ACF_LAGS = 60
SCATTER_SAMPLES = 20000
REPRESENTATIVE_SAMPLE_COUNT = 6
```

```python
import pandas as pd
from IPython.display import Image, display

from Coursework3.RNO_1D_Skeleton import run_eda_and_preprocessing

eda_results = run_eda_and_preprocessing(
    TRAIN_PATH=TRAIN_PATH,
    ARTIFACT_ROOT=ARTIFACT_ROOT,
    F_FIELD=F_FIELD,
    SIG_FIELD=SIG_FIELD,
    SEED=SEED,
    ACF_LAGS=ACF_LAGS,
    SCATTER_SAMPLES=SCATTER_SAMPLES,
    REPRESENTATIVE_SAMPLE_COUNT=REPRESENTATIVE_SAMPLE_COUNT,
)

summary_df = pd.DataFrame(eda_results['summary_rows'])
split_df = pd.DataFrame(eda_results['split_rows'])
normalization_df = pd.DataFrame(eda_results['normalization_rows'])
stationarity_df = pd.DataFrame(eda_results['stationarity_rows'])

display(summary_df)
display(split_df)
display(normalization_df)
display(stationarity_df)
```
