# 04 Inference And Testing

Evaluate the final checkpoint on the untouched test set and run qualitative unseen-load inference.

```python
import json
from pathlib import Path

ARTIFACT_ROOT = Path('artifacts')
SEED = 20260328
USE_OPTUNA_BEST = True
CHECKPOINT_PATH = ARTIFACT_ROOT / 'checkpoints' / 'final_model.pt'

if USE_OPTUNA_BEST:
    best_param_paths = sorted((ARTIFACT_ROOT / 'optuna').glob('best_params_*.json'))
    best_payload = None
    for path in best_param_paths:
        payload = json.loads(path.read_text())
        if best_payload is None or payload['best_value'] < best_payload['best_value']:
            best_payload = payload
    if best_payload is None:
        raise FileNotFoundError('No Optuna best-parameter files found.')
    CHECKPOINT_PATH = Path(best_payload['best_user_attrs']['checkpoint_path'])
```

```python
import pandas as pd
from IPython.display import Image, display

from Coursework3.RNO_1D_Skeleton import run_inference_and_testing

inference_results = run_inference_and_testing(
    checkpoint_path=CHECKPOINT_PATH,
    artifact_root=ARTIFACT_ROOT,
    split_seed=SEED,
)

pd.DataFrame([inference_results['metrics']])
```
